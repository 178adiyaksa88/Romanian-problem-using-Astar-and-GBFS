```python
df = df[['id', 'NObeyesdad', 'Age', 'Height', 'Weight', 'FCVC', 'NCP', 'CH2O', 'FAF', 'TUE', 'Gender_Male',
     'family_history_with_overweight_yes', 'FAVC_yes', 'CAEC_Frequently', 'CAEC_Sometimes', 'CAEC_no', 'SMOKE_yes',
     'SCC_yes', 'CALC_Frequently', 'CALC_Sometimes', 'CALC_no', 'MTRANS_Bike', 'MTRANS_Motorbike',
     'MTRANS_Public_Transportation', 'MTRANS_Walking']]

df.head()
```

```python
scaler = StandardScaler()
df.iloc[:,2:10] = scaler.fit_transform(df.iloc[:,2:10])
```

```python
train_df = df.iloc[:20758, :]   # train data set
test_df = df.iloc[20758:, :]    # test data set

y = train_df[["NObeyesdad"]]
X = train_df.drop(["NObeyesdad", "id"], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python

from sklearn.svm import SVC  # Import SVM

# Updated model list including SVM
model_list = [ SVC()]

model_name_list = []
model_accuracy_list = []

for model in model_list:
    model_name = model.__class__.__name__

    if model_name == 'SVC':  # If the model is SVM, consider different parameters
        model = SVC(kernel='rbf')  # You can adjust kernel and other parameters as needed
    else:
        model = model

    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)

    model_name_list.append(model_name)
    model_accuracy_list.append(accuracy)

    print(f"{model_name} accuracy: {accuracy:.3f}\n")
```
